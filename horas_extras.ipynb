```python
def estandarizar_he (ruta, mes, periodo):
    df = pd.read_excel(ruta, sheet_name=mes, header=1)
    df1 = df[["SEDE", "SEC", "DNI", "PUESTO", "CONDICION", 'SALDO DEL PERIODO', 'TOTAL COMPENSACION', 'HE DEL PERIODO + SALDO INICIAL', 'TOTAL HORAS A PAGAR']]
    df1["PERIODO"] = periodo
    display(df1.head(1))
    return df1
```

```python
dff1 = dff.merge(df_dic[["SEDE", "codigo_local"]], on="SEDE", how="left")
```

```python
dff1[dff1["PERIODO"].isin(["202501", "202502", "202503", "202504", "202601", "202602", "202603", "202604"])].groupby(["codigo_local", "AÑO"])["TOTAL HORAS A PAGAR"].sum().reset_index()
```

```python
dff2 = dff1[dff1["PERIODO"].isin(["202501", "202502", "202503", "202601", "202602", "202603"])].pivot_table(index="codigo_local",
    columns="AÑO",
    values="TOTAL HORAS A PAGAR",
    aggfunc="sum",  # Evita duplicidad si hay múltiples registros
).reset_index()
```
