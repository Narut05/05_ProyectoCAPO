```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from glob import glob
# Aplicar configuraciones de visualización total de Pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)

from pathlib import Path
import glob
```

```python

path_carpeta = Path("data")
archivos_csv = glob.glob(str(path_carpeta / "*.csv"))

df = pd.concat( [pd.read_csv(archivo) for archivo in archivos_csv], ignore_index=True)
```

```python
df1 = df.dropna(subset="Export")
```

```python
df2 = df1[df1["Área"]!="SIN ELECTRO"]
```

```python
df2["Área"].unique()
```

```python
df2["Periodo"] = df2["Periodo"].astype(int)
df2["Periodo"] = df2["Periodo"].astype(str)
```

```python
df2["Tienda"].unique()
```

```python
df2["Year"] = df2["Periodo"].str[0:4]
df2["Mes"] = df2["Periodo"].str[4:6]

df3 = df2[~df2["Tienda"].isin(["V019 Pardo - VIV", "X209 Bolognesi - PVE", "X210 Santa Cruz - PVE", "Z991 Back Office Plaza Vea", "Z992 Back Office Vivanda", "Z993 Retail Media - Pv", "V254 Destinos del Sabor - VIV"])]
df3
```

```python
area = "CAJAS"
tienda = '1199 San Miguel - PVH'
dff = df3[(df3["Área"]==area)&(df3["Mes"].isin(["03", "04", "05"]))][["Tienda", "Year", "Mes", "Área", "Periodo", "Prod", "VtaNeta", "JEq", "CAPO", "CAPE"]].sort_values(by=["Tienda","Periodo"])
# dff = df2[(df2["Área"]==area)].groupby(["Periodo"]).agg({"Prod":"sum", "VtaNeta":"sum", "JEq":"sum", "CAPO":"sum", "CAPE":"sum"}).reset_index().sort_values(by="Periodo")
dff["ratio_jeq_capo"] =  dff["JEq"] / dff["CAPO"]
dff["ratio_jeq_cape"] =  dff["JEq"] / dff["CAPE"]
dff["ratio_capo_cape"] = dff["CAPE"] / dff["CAPO"]

dff["ratio_jeq_capo_year"] = dff.groupby(["Tienda", "Year"])["JEq"].transform("sum") / dff.groupby(["Tienda", "Year"])["CAPO"].transform("sum")
dff["ratio_jeq_cape_year"] = dff.groupby(["Tienda", "Year"])["JEq"].transform("sum") / dff.groupby(["Tienda", "Year"])["CAPE"].transform("sum")
dff["ratio_capo_cape_year"] = dff.groupby(["Tienda", "Year"])["CAPE"].transform("sum") / dff.groupby(["Tienda", "Year"])["CAPO"].transform("sum")
dff

dff1 = dff.groupby(["Tienda", "Year"]).agg({"ratio_jeq_capo_year":"mean", "ratio_jeq_cape_year":"mean", "ratio_capo_cape_year":"mean"}).reset_index()
df_pivot = dff1.pivot_table(
    index="Tienda",  # Variable que irá en las filas
    columns="Year",  # Variable que se abrirá en las columnas
    values=["ratio_jeq_capo_year", "ratio_jeq_cape_year", "ratio_capo_cape_year"],  # Métrica numérica a agregar
    aggfunc="sum",  # Operación estadística (mean, sum, count, median)
).reset_index()


# df_pivot.columns
# # 1. Definición de vectores desde la estructura Multi-Index
# r_2024 = df_pivot[("ratio_jeq_capo_year", "2024")]
# r_2025 = df_pivot[("ratio_jeq_capo_year", "2025")]
# r_2026 = df_pivot[("ratio_jeq_capo_year", "2026")]

# # 2. Parametrización de la banda de tolerancia (+/- 5%) respecto al baseline (2025)
# baseline_historico = (r_2024 + r_2025) / 2.0
# tol_superior = baseline_historico * 1.03
# tol_inferior = baseline_historico * 0.93

# # 3. Lógica condicional vectorizada
# # Las desviaciones que no superen estos umbrales caen automáticamente en el 'default' (Igual)
# condiciones = [
#     r_2026 > tol_superior,  # Incremento real significativo
#     r_2026 < tol_inferior,  # Decremento real significativo
# ]

# resultados = ["Mejor", "Peor"]

# # Asignación directa a la nueva columna del Pivot Table
# df_pivot[("Conclusion", "ratio_jeq_capo")] = np.select(
#     condiciones, resultados, default="Igual (Estable dentro del +/-3%)"
# )

# df_pivot.tail(10)
```

```python
# 1. Definición del mapeo de métricas base y sus respectivas salidas de conclusión
mapeo_metricas = {
    "ratio_jeq_capo_year": "ratio_jeq_capo",
    "ratio_jeq_cape_year": "ratio_jeq_cape",
    "ratio_capo_cape_year": "ratio_capo_cape",
}

# 2. Configuración de etiquetas de año (ajustar a int si tu DataFrame usa enteros: 2024, 2025, 2026)
y24, y25, y26 = "2024", "2025", "2026"

# 3. Procesamiento vectorial iterativo para inyectar las columnas en el Multi-Index
for metric_sub, conclusion_name in mapeo_metricas.items():
    # Extracción de vectores por año
    r_2024 = df_pivot[(metric_sub, y24)]
    r_2025 = df_pivot[(metric_sub, y25)]
    r_2026 = df_pivot[(metric_sub, y26)]

    # Baseline: Promedio bianual (2024-2025)
    baseline_historico = (r_2024 + r_2025) / 2.0

    # Umbrales con banda de tolerancia del +/- 3%
    tol_superior = baseline_historico * 1.03
    tol_inferior = (
        baseline_historico * 0.97
    )  # Corregido de 0.93 a 0.97 para mantener el 3%

    # Máscaras booleanas vectorizadas
    condiciones = [r_2026 > tol_superior, r_2026 < tol_inferior]
    resultados = ["Mejor", "Peor"]

    # Asignación directa en la estructura de columnas jerárquicas
    df_pivot[("Conclusion", conclusion_name)] = np.select(
        condiciones, resultados, default="Igual (Estable dentro del +/-3%)"
    )
```
