```python
import json
import geopandas as gpd
import pandas as pd


def normalize_departments(series: pd.Series) -> pd.Series:
    return (
        series.astype(str)
        .str.upper()
        .str.normalize("NFKD")
        .str.encode("ascii", errors="ignore")
        .str.decode("utf-8")
        .str.strip()
    )


# Opción A: Si tienes el archivo guardado en el disco (Recomendado)
gdf_peru = gpd.read_file("mapa_depart.json")

# Opción B: Si necesitas meter el texto de la captura en una variable temporal por única vez
# json_data = """ <pega aquí el bloque de texto de tu captura> """
# gdf_peru = gpd.GeoDataFrame.from_features(json.loads(json_data)["features"])

# Estandarizar la clave de cruce geográfico
gdf_peru["NOMBDEP_clean"] = normalize_departments(gdf_peru["NOMBDEP"])
```
